```python
#Atividade 01 - Classificar a coluna localização de acordo com a classificação CDU.

CDU_lista = []
for CDU in dados_completos_biblioteca['localizacao']:
  if (CDU <100):
    CDU_lista.append('Generalidades')
  elif (CDU <200):
    CDU_lista.append('Filosifia. psicologia')
  elif (CDU <300):
    CDU_lista.append('Religião')
  elif (CDU <400):
    CDU_lista.append('Ciência sociais')
  elif (CDU <500):
    CDU_lista.append('Classe Vaga')
  elif (CDU <600):
    CDU_lista.append('Ciência sociais')
  elif (CDU <700):
    CDU_lista.append('Belas artes')
  elif (CDU <800):
    CDU_lista.append('Linguagem')
  else:
    CDU_lista.append('Geografia')

dados_completos_biblioteca['CD_Geral'] = CDU_lista
```

# Terceiro dia

O Cliente solicitou as seguintes analises:

Visualização da quantidade de totais de emprestimos por ano. Obs: Exemplares emprestados, e não de empréstimos realizados.

O Cliente gostaria de visualizar os meses com a menor demanda de emprestimos. Com essas visualização ele vai conseguir tomar a desição de alocação de recursoss humanos. Existe uma sentimento que os meses de Março e Semtebro são os maiores.

O Cliente gostaria de visualizar os dados em um grafico em Barra para os meses com o maiores fluxos de emprestimos.







```python
dados_completos_biblioteca.head(2)
```

```python
#Novo DataFreme para analise e manipulação dos Dados para demanda do dia 03
analise_emprestimos = dados_completos_biblioteca
```

```python
dados_emprestimos.head(2)
```

```python
#Converte datas para analise de emprestimos x Ano
analise_emprestimos['data_emprestimo'] = pd.to_datetime(analise_emprestimos['data_emprestimo'])
analise_emprestimos['data_renovacao'] = pd.to_datetime(analise_emprestimos['data_renovacao'])
analise_emprestimos['data_devolucao'] = pd.to_datetime(analise_emprestimos['data_devolucao'])
```

```python
#Adicionando uma coluna chamada Ano para analise dos emprestimos por ano
analise_emprestimos['Ano'] = analise_emprestimos['data_emprestimo'].dt.year
```

```python
#Agrupanos os dados pela coluna ano
emprestimos_ano = analise_emprestimos.groupby('Ano')['id_emprestimo'].count()
```

```python
#Ordenar emprestimos do maior para menor
emprestimos_ano = emprestimos_ano.sort_values(ascending=False)
emprestimos_ano
```

```python
emprestimos_ano = emprestimos_ano.rename("Quantidade")
emprestimos_ano
```

```python
#Trabalhando com grafico
%matplotlib inline
import matplotlib.pyplot as plt

#criando grafico por barra
emprestimos_ano.plot(kind='bar')
#Definir os rotulos do eixo x e y
plt.xlabel('Ano')
plt.ylabel('Quantidade de Emprestimos')
#definir o titulo para o grafico
plt.title('Quantidade de Emprestimos por Ano')
#exibir o grafico
plt.show()
```

```python
#Dia 03 - Analise dos mêses com o maior quantidade de emprestimos
#Adicionando uma coluna mês 'dt.month'
analise_emprestimos['Mes'] = analise_emprestimos['data_emprestimo'].dt.month
```

```python
#Adicionando um grupo para analisar a quantidade por Mês e ordenar pelos maior valores
emprestimos_mes = analise_emprestimos.groupby('Mes')['id_emprestimo'].count()
emprestimos_mes = emprestimos_mes.sort_values(ascending=False)
emprestimos_mes
```

```python
#Renomendo os números dos meses para o nome

emprestimos_mes = emprestimos_mes.rename(index=calendar.month_name)
emprestimos_mes
```

```python
#Adicionando Grafico dos emprestimos por mês
ax = emprestimos_mes.plot(kind='bar')
#definir os rotulos para x e y
ax.set_xlabel('Mês')
ax.set_ylabel('Quantidade de Empréstimos')
#definir o título
ax.set_title('Maior quantidade de Emprestimos por Mês')
#incluindo os nomes dos meses no grafico
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
#exibir o grafico
plt.show()
```
